# transparency

```python
a=np.array([6,2,8,3,7,3,6,13])
b=np.array([7,1,5,4,2,8,12,14])
colors=np.array(["red", "green", "blue", "yellow", "purple", "orange", "pink", "brown"])
sizes=np.array([120, 220, 123, 422, 215, 346, 237, 128])
plt.scatter(a,b,c=colors,s=sizes,alpha=0.2)          #use alpha 0-2
plt.colorbar()
```

```python
a=np.array([6,2,8,3,7,3,6,13])
b=np.array([7,1,5,4,2,8,12,14])
colors=np.array(["red", "green", "blue", "yellow", "purple", "orange", "pink", "brown"])
sizes=np.array([120, 220, 123, 422, 215, 346, 237, 128])
plt.scatter(a,b,c=colors,s=sizes,cmap='nipy_spectral')
plt.colorbar()
plt.colorbar(label="color map")
```

# for Piechart

```python
y=np.array([30,11,80,90])
mylabels=["apple","guava","banana","watermelon"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,startangle=0,explode=myexplode)  #startangle
plt.show()
```

```python
y=np.array([30,11,80,90])
mylabels=["apple","guava","banana","mango"]
myexplode=[0.2,0.3,0.2,0.02]
color=np.array(["red", "green", "yellow", "orange"])
plt.pie(y,labels=mylabels,startangle=0,explode=myexplode,shadow=True,colors=color)  #startangle # myexplode = gap 
plt.show()
```

# legend()

```python
y=np.array([30,11,80,90])
mylabels=["apple","guava","banana","watermelon"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,startangle=0,explode=myexplode)  #startangle
plt.legend(title="fruit names") #title="fruit names"
plt.show()
```

```python
data=pd.read_csv("Used_Bikes.csv")
import matplotlib.pyplot as plt
plt.scatter(data['price'],data['brand'])
```

```python
data = pd.read_csv('company_sales_data.csv')
plt.plot(data['month_number'],data['total_profit'])
plt.plot(data['total_units'])
plt.legend("mt")
plt.show()
```

```python
data = pd.read_csv('company_sales_data.csv')
plt.plot(data['month_number'],data['bathingsoap'],marker='o',linewidth=1,label="bathing soap")
plt.plot(data['month_number'],data['facecream'],marker='o',linewidth=1,label="face cream")
plt.plot(data['month_number'],data['facewash'],marker='o',linewidth=1,label="facewash")
plt.plot(data['month_number'],data['toothpaste'],marker='o',linewidth=1,label="toothpaste")
plt.plot(data['month_number'],data['shampoo'],marker='o',linewidth=1,label="shampoo")
plt.plot(data['month_number'],data['moisturizer'],marker='o',linewidth=1,label="moisturizer")
plt.legend()
plt.show()
```

```python
plt.scatter(data['month_number'],data['toothpaste'],marker='o',linewidth=1,label="toothpaste")
plt.grid(linestyle='--')
plt.title("Toothpaste sales data")
plt.xlabel("Month Number")
plt.ylabel("Number of Units Sold")
plt.legend()
```
